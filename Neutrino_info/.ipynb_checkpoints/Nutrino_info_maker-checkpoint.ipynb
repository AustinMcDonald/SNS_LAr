```python
import numpy as np
import matplotlib.pyplot as plt

from scipy import interpolate
from scipy import integrate

import pandas as pd
```

```python
X_sections = "Neutrino_Xsections.csv"
```

```python
X_SEC = pd.read_csv(X_sections)
```

```python
X_SEC
```

```python
FAC = 1e-38

IBD_X = X_SEC["IBD"][1:].to_numpy().astype(np.float)
IBD_X = IBD_X[~np.isnan(IBD_X)]
IBD_Y = X_SEC["Unnamed: 1"][1:].to_numpy().astype(np.float)* FAC
IBD_Y = IBD_Y[~np.isnan(IBD_Y)]

nue_e_X = X_SEC["nue_e-"][1:].to_numpy().astype(np.float)
nue_e_X = nue_e_X[~np.isnan(nue_e_X)]
nue_e_Y = X_SEC["Unnamed: 3"][1:].to_numpy().astype(np.float)* FAC
nue_e_Y = nue_e_Y[~np.isnan(nue_e_Y)]

nueb_e_X = X_SEC["nueb_e-"][1:].to_numpy().astype(np.float)
nueb_e_X = nueb_e_X[~np.isnan(nueb_e_X)]
nueb_e_Y = X_SEC["Unnamed: 5"][1:].to_numpy().astype(np.float)* FAC
nueb_e_Y = nueb_e_Y[~np.isnan(nueb_e_Y)]

nux_e_X = X_SEC["nux_e-"][1:].to_numpy().astype(np.float)
nux_e_X = nux_e_X[~np.isnan(nux_e_X)]
nux_e_Y = X_SEC["Unnamed: 7"][1:].to_numpy().astype(np.float)* FAC
nux_e_Y = nux_e_Y[~np.isnan(nux_e_Y)]

nuxb_e_X = X_SEC["nuxb_e-"][1:].to_numpy().astype(np.float)
nuxb_e_X = nuxb_e_X[~np.isnan(nuxb_e_X)]
nuxb_e_Y = X_SEC["Unnamed: 9"][1:].to_numpy().astype(np.float)* FAC
nuxb_e_Y = nuxb_e_Y[~np.isnan(nuxb_e_Y)]

CNC_X = X_SEC["conNC-p"][1:].to_numpy().astype(np.float)
CNC_X = CNC_X[~np.isnan(CNC_X)]
CNC_Y = X_SEC["Unnamed: 11"][1:].to_numpy().astype(np.float)* FAC
CNC_Y = CNC_Y[~np.isnan(CNC_Y)]

nue_Ar_X = X_SEC["nue_Ar"][1:].to_numpy().astype(np.float)
nue_Ar_X = nue_Ar_X[~np.isnan(nue_Ar_X)]
nue_Ar_Y = X_SEC["Unnamed: 13"][1:].to_numpy().astype(np.float)* FAC
nue_Ar_Y = nue_Ar_Y[~np.isnan(nue_Ar_Y)]
```

```python
plt.figure(figsize=(10,8))
plt.plot(IBD_X, IBD_Y, label="IBD", lw=2)
plt.plot(nue_e_X, nue_e_Y, label="nue_e", lw=2)
plt.plot(nueb_e_X, nueb_e_Y, label="nueb_e", lw=2)
plt.plot(nux_e_X, nux_e_Y, label="nux_e", lw=2)
plt.plot(nuxb_e_X, nuxb_e_Y, label="nuxb_e", lw=2)
plt.plot(CNC_X, CNC_Y, label="CNC", lw=2)
plt.plot(nue_Ar_X, nue_Ar_Y, label="nue_Ar", lw=2)

plt.legend(loc='upper left',ncol=3,fontsize=18)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(which="both",linestyle='--')
plt.xlabel('Energy [MeV]',fontsize=18)
plt.ylabel(r'Cross-section [cm$^2$]',fontsize=18)

plt.semilogy()
plt.xlim(5,100)
plt.ylim(4e-45, 4e-37)
plt.show()
```

```python
Energy = np.linspace(1,101,num=500,endpoint=False)
```

```python
IBD_I      = interpolate.interp1d(IBD_X, IBD_Y,  kind='linear', fill_value="extrapolate")

nue_e_I    = interpolate.interp1d(nue_e_X, nue_e_Y,  kind='linear', fill_value="extrapolate")
nueb_e_I   = interpolate.interp1d(nueb_e_X, nueb_e_Y,  kind='linear', fill_value="extrapolate")

nux_e_I    = interpolate.interp1d(nux_e_X, nux_e_Y,  kind='linear', fill_value="extrapolate")
nuxb_e_I   = interpolate.interp1d(nuxb_e_X, nuxb_e_Y,  kind='linear', fill_value="extrapolate")

CNC_I      = interpolate.interp1d(CNC_X, CNC_Y,  kind='linear', fill_value="extrapolate")
nue_Ar_I   = interpolate.interp1d(nue_Ar_X, nue_Ar_Y,  kind='linear', fill_value="extrapolate")
```

```python
plt.figure(figsize=(10,8))
plt.plot(Energy, IBD_I(Energy), label="IBD", lw=3)
plt.plot(Energy, nue_e_I(Energy), label=r"$\nu_e$-e", lw=3)
plt.plot(Energy, nueb_e_I(Energy), label=r"$\overline{\nu}_e$-e", lw=3)
plt.plot(Energy, nux_e_I(Energy), label=r"$\nu_x$-e", lw=3)
plt.plot(Energy, nuxb_e_I(Energy), label=r"$\overline{\nu}_x$-e", lw=3)
plt.plot(Energy, CNC_I(Energy), label="CNC", lw=3)
plt.plot(Energy, nue_Ar_I(Energy), label=r"$\nu_e$-Ar", lw=3)

plt.legend(loc='upper left',ncol=3,fontsize=18)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(which="both",linestyle='--')
plt.xlabel('Energy [MeV]',fontsize=18)
plt.ylabel(r'Cross-section [cm$^2$]',fontsize=18)

plt.semilogy()
plt.xlim(1,100)
plt.ylim(4e-45, 4e-37)
plt.tight_layout()
plt.savefig("cross-sections.pdf")
plt.show()
```

```python
## Imported data from Figure 01 of https://arxiv.org/pdf/1211.5199.pdf ###
## Electron Neutrino Flux in units of Neutrinos / cm^2 / day as a function of energy (MeV)
ElectronNeutrino_DataFile = 'SNS-electronNeutrino-flux.csv'
MuonAntiNeutrino_DataFile = 'SNS-antimuonneutrino-flux.csv'
```

```python
### Loading the data into x & y variables I can manipulate ###
ElectronNu_data = np.loadtxt(ElectronNeutrino_DataFile, delimiter=',')
x_ENu_Energy = ElectronNu_data[:,0]
y_ENu_Flux = ElectronNu_data[:,1]

AntiMuonNu_data = np.loadtxt(MuonAntiNeutrino_DataFile, delimiter=',')
x_aMuNu_Energy = AntiMuonNu_data[:,0]
y_aMuNu_Flux = AntiMuonNu_data[:,1]

x_MuNu_Energy = np.array([30,30])
y_MuNu_Flux = np.array([0,4700*1e6])
```

```python
plt.figure(figsize=(10,8))
plt.plot(x_ENu_Energy, y_ENu_Flux, label="Nu_e", lw=2)
plt.plot(x_aMuNu_Energy, y_aMuNu_Flux, label="Nu_mub", lw=2)

plt.plot(x_MuNu_Energy, y_MuNu_Flux, label="Nu_mu", lw=2)

plt.legend(loc='upper left',ncol=1,fontsize=18)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(which="both",linestyle='--')
plt.xlabel('Energy [MeV]',fontsize=18)
plt.ylabel(r'Flux ',fontsize=18)

# plt.semilogy()
# plt.xlim(5,100)
# plt.ylim(4e-45, 4e-37)
plt.show()
```

```python
F_numub   = interpolate.interp1d(x_aMuNu_Energy, y_aMuNu_Flux,  kind='linear', fill_value=(0,0), bounds_error=False)
F_nue     = interpolate.interp1d(x_ENu_Energy, y_ENu_Flux,  kind='linear', fill_value=(0,0), bounds_error=False)
```

```python
loc = np.where(Energy==30)[0][0]
F_numu = np.zeros(len(Energy))
F_numu[loc] = 4700*1e6
```

```python
plt.figure(figsize=(10,8))
plt.plot(Energy, F_numub(Energy), label=r"$\overline{\nu}_\mu$", lw=3)
plt.plot(Energy, F_nue(Energy), label=r"${\nu}_e$", lw=3)
plt.plot(Energy, F_numu, label=r"${\nu}_\mu$", lw=3)


plt.legend(loc='upper left',ncol=1,fontsize=18)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(which="both",linestyle='--')
plt.xlabel('Energy [MeV]',fontsize=18)
plt.ylabel(r'Flux [neutrinos / 0.2MeV / cm$^2$]',fontsize=18)

# plt.semilogy()
plt.xlim(0,60)
# plt.ylim(4e-45, 4e-37)
plt.tight_layout()
plt.savefig("Flux.pdf")
plt.show()
```

```python
class Neutrino_Info:
    def __init__(self):
        """
        Fill the empty variables. 
        """
        self.Energy     = np.array([])
        
        self.Flux_numu  = np.array([])
        self.Flux_numub = np.array([])
        self.Flux_nue   = np.array([])
        
        self.XS_IBD     = np.array([])
        self.XS_nue_e   = np.array([])
        self.XS_nueb_e  = np.array([])
        self.XS_nux_e   = np.array([])
        self.XS_nuxb_e  = np.array([])
        self.XS_CNC     = np.array([])
        self.XS_nue_Ar  = np.array([])
```

```python
Nu_Info = Neutrino_Info()
```
