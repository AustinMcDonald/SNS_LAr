```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
fXSection = '/Users/yuntse/work/sns/SNS_LAr/Neutrino_info/Neutrino_Xsections.csv'
```

```python
XSection = pd.read_csv(fXSection, header=1, names = [ 'IBDE', 'IBD', 'nue_e-E', 'nue_e-', 'nueb_e-E', 'nueb_e-', 'nux_e-E', 'nux_e-', 'nuxb_e-E', 'nuxb_e-', 'conNC-pE', 'conNC-p', 'nue_ArE', 'nue_Ar'])
```

```python
XSection
```

```python
XSection['IBD']
```

```python
FAC = 1e-38
```

```python
plt.figure(figsize = (10,8))
plt.plot(XSection['IBDE'], XSection['IBD']*FAC, label = 'IBD', lw = 2)
plt.plot(XSection['nue_e-E'], XSection['nue_e-']*FAC, label = r'$\nu_e$-e', lw = 2)
plt.plot(XSection['nueb_e-E'], XSection['nueb_e-']*FAC, label = r'$\overline{\nu}_e$-e', lw = 2)
plt.plot(XSection['nux_e-E'], XSection['nux_e-']*FAC, label = r'$\nu_x$-e', lw = 2)
plt.plot(XSection['nuxb_e-E'], XSection['nuxb_e-']*FAC, label = r'$\overline{\nu}_x$-e', lw = 2)
plt.plot(XSection['conNC-pE'], XSection['conNC-p']*FAC, label = 'CNC', lw = 2)
plt.plot(XSection['nue_ArE'], XSection['nue_Ar']*FAC, label = r'$\nu_e$-Ar', lw = 2)

plt.legend(loc='upper left', ncol = 3, fontsize = 18)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.grid(which = "both", linestyle = '--')
plt.xlabel('Energy [MeV]', fontsize = 18)
plt.ylabel(r'Cross-section [cm$^2$]', fontsize = 18)

plt.semilogy()
plt.xlim(5, 100)
plt.ylim(4e-45, 4e-37)
plt.show()
```
